```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("screen.jpg")
```

```python
def show(img):
    plt.axis("off")
    plt.imshow(image)
    plt.show()
```

```python
lower = np.array((0, 0, 200), dtype = "uint8")
upper = np.array((20, 20, 255), dtype = "uint8")
```

```python
mask = cv2.inRange(image, lower, upper)
output = cv2.bitwise_and(image, image, mask = mask)
show(output)
```
